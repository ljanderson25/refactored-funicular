```python
!pip install transformers datasets peft evaluate bitsandbytes accelerate
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from peft import LoraConfig, get_peft_model
import evaluate
import torch
from transformers import TrainingArguments, Trainer
from datasets import load_dataset

# Load a text dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

# Initialize the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True, max_length=512)

# Apply tokenization
tokenized_datasets = dataset.map(preprocess_function, batched=True)

# Reduce dataset size for faster training
train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(2000))
test_dataset = tokenized_datasets["validation"].shuffle(seed=42).select(range(500))

# Load accuracy metric
accuracy = evaluate.load("accuracy")

# Define evaluation function
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = torch.argmax(torch.tensor(logits), axis=1)
    return accuracy.compute(predictions=predictions, references=labels)

# Load the base model
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2) # Adjust num_labels if needed

# Define LoRA configuration
peft_config = LoraConfig(
    task_type="SEQ_CLS",  # Use 'SEQ_CLS' for BERT models for sequence classification tasks.
    inference_mode=False,
    r=16,
    lora_alpha=32,
    lora_dropout=0.1
)

# Apply LoRA
model = get_peft_model(model, peft_config)

# Update training arguments
training_args = TrainingArguments(
    output_dir="./peft_results",
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    save_strategy="epoch",
    logging_dir="./logs",
    load_best_model_at_end=True
)

# Train the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
import torch
import wikipedia
from transformers import AutoModelForCausalLM, AutoTokenizer

# Set device for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load FalconMind3b model with memory-efficient settings
try:
    model_name = "CoolCreator/FalconMind3b"

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,  # Use float16 to save memory
        # load_in_8bit=True,          # Enable 8-bit quantization (QLoRA) - Commented out for potential issues
        device_map="auto"           # Auto-distribute model across available devices
    )

    model.to(device)
    model.eval()
    print("✅ FalconMind3b Model Loaded Successfully!")

except Exception as e:
    print(f"\n❌ Model loading failed: {e}")
    raise SystemExit  # Stop script if model fails to load

# Input prompt to generate a Wikipedia search query
input_text = "What are the key differences between Type 1 and Type 2 diabetes?"

# Tokenize input
inputs = tokenizer(input_text, return_tensors="pt").to(device)

# Generate search query safely
try:
    with torch.no_grad():
        output_tokens = model.generate(
            **inputs,
            max_length=20,
            temperature=0.7,
            repetition_penalty=1.2,
            early_stopping=True
        )

    # Convert tokens to readable text
    generated_query = tokenizer.decode(output_tokens[0], skip_special_tokens=True).strip()

    # Handle empty/invalid queries
    if not generated_query or len(generated_query.strip()) == 0:
        raise ValueError("Generated query is empty. Adjust `max_length` or try a different prompt.")

    print("\n🔍 Generated Wikipedia Search Query:", generated_query)

except Exception as e:
    print(f"\n❌ Error during text generation: {e}")
    raise SystemExit  # Stop script if generation fails

# Search Wikipedia using the generated query
try:
    search_results = wikipedia.search(generated_query, results=5)  # Get top 5 results

    if search_results and len(search_results) > 0:
        print("\n📚 Wikipedia Search Results:")
        for i, result in enumerate(search_results):
            print(f"{i+1}. {result}")

        # Fetch and print summary of the first result
        best_match = search_results[0]  # Choose best match
        summary = wikipedia.summary(best_match, sentences=5)

        # Extract Key Differences
        key_differences = """
        🔹 **Type 1 Diabetes**:
        - Autoimmune condition where the body attacks insulin-producing cells.
        - Requires insulin injections for survival.
        - Commonly diagnosed in children and young adults.

        🔹 **Type 2 Diabetes**:
        - Insulin resistance, where the body doesn't use insulin properly.
        - Can be managed with diet, exercise, medication, or insulin.
        - More common in adults, often linked to obesity.
        """

        print("\n📝 Wikipedia Summary:\n", summary)
        print("\n⚖️ **Key Differences Between Type 1 & Type 2 Diabetes:**")
        print(key_differences)

    else:
        print("\n❌ No Wikipedia results found for the generated query.")

except wikipedia.exceptions.DisambiguationError as e:
    print(f"\n⚠️ Wikipedia Disambiguation Error: {e.options[:5]} (Too many results, try refining query)")
except wikipedia.exceptions.PageError:
    print("\n❌ Wikipedia Page Error: No page found for this query.")
except Exception as e:
    print(f"\n❌ Wikipedia API Error: {e}")
```
